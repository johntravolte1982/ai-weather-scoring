```python
import pandas as pd
import matplotlib.pyplot as plt

# Загружаем данные
metar = pd.read_csv('data/raw/metar_historical.csv', index_col=0, parse_dates=True)
ecmwf = pd.read_csv('data/raw/ecmwf_forecast.csv', index_col=0, parse_dates=True)

# Объединяем по времени
merged = metar.join(ecmwf, how='inner')

# Смотрим расхождение
merged['temp_error'] = merged['temp_c'] - merged['ecmwf_temp_c']

print(f"Средняя ошибка ECMWF: {merged['temp_error'].mean():.2f}°C")
print(f"RMSE: {((merged['temp_error'] ** 2).mean()) ** 0.5:.2f}°C")

# График
plt.figure(figsize=(12, 6))
plt.plot(merged.index, merged['temp_c'], label='METAR (факт)', alpha=0.7)
plt.plot(merged.index, merged['ecmwf_temp_c'], label='ECMWF (прогноз)', alpha=0.7)
plt.legend()
plt.title('Сравнение прогноза ECMWF с фактической погодой')
plt.grid(True, alpha=0.3)
plt.show()
```
